```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display,display_html
from operator import itemgetter
df= pd.read_csv('AIS datasheet.csv')
listOfNamesOfNDC=['PHRMCY_NAME','BUY_GRP_NAM','NDC_DESC','DSPN_QTY_NDC_1WK','PURCH_QTY_NDC_1WK_MTRC','DSPN_QTY_NDC_4WKS','PURCH_QTY_NDC_4WKS_MTRC','DSPN_QTY_NDC_13WKS','PURCH_QTY_NDC_13WKS_MTRC','DSPN_QTY_NDC_26WKS','PURCH_QTY_NDC_26WKS_MTRC','DSPN_QTY_GCN_1WK','PURCH_QTY_GCN_1WK_MTRC','DSPN_QTY_GCN_4WKS','PURCH_QTY_GCN_4WKS_MTRC','DSPN_QTY_GCN_13WKS','PURCH_QTY_GCN_13WKS_MTRC','DSPN_QTY_GCN_26WKS','PURCH_QTY_GCN_26WKS_MTRC']
df1= df[listOfNamesOfNDC]
#X= df1.values
dictFinal=dict()
dictNew=dict()
index=0
dfNew=df1['NDC_DESC'].drop_duplicates(inplace=False)
#print(dfNew)
for each in dfNew:
    dictNew[index] = each
    index+=1
#print(dictNew)
k=0
#print(dfNew[1])
while k < index:
    listN=[]
   # print(X.shape)
    for each in df1.values:
        #print(each)
        #print(each[2],':  Yessss  :',dfNew[k])
        if each[2] == dictNew[k]:
            listN.append(each)
            
    #print(dfNew[k],len(listN))
    a= np.array(listN)
    #print(a)
    #d= pd.DataFrame.from_records(a)
    e=pd.DataFrame(a,columns=listOfNamesOfNDC)
    X= e.iloc[:,2:20]
    isPVA=0
    #display(X)
    for i in range(0,X.shape[0]):
        count=0
        temp= X.iloc[i,:].values
        for i in range(1,len(temp),2):
            if abs(temp[i]-temp[i+1]) >50:
                count+=1
        if count ==0:
            isPVA+=1
    dictFinal[isPVA] = dictNew[k]
    #print(dictNew[k],isPVA,X.shape[0])
    k+=1
a=sorted(dictFinal.items(),key= itemgetter(0),reverse=True)
keys=[]
values=[]
for each in a:
    keys.append(each[0])
    values.append(each[1])
dt1 = pd.DataFrame(keys,columns=['PVA_COUNT_OF_EACH_DRUG_FAMILY'])
dt2 = pd.DataFrame(values,columns=['NAME_OF_DRUG_FAMILY'])
dt1Style= dt1.style.set_table_attributes("style='display:inline'").set_caption('Table 1')
dt2Style= dt2.style.set_table_attributes("style='display:inline'").set_caption('Table 2')
display_html(dt1Style._repr_html_() + dt2Style._repr_html_(), raw=True)
plt.figure(figsize=(16,6))
sns.barplot(x=dt2['NAME_OF_DRUG_FAMILY'][0:5],y=dt1['PVA_COUNT_OF_EACH_DRUG_FAMILY'][0:5],palette='spring')
plt.show()
```
